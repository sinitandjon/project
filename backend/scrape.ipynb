```python
import requests
from bs4 import BeautifulSoup
import time


# URL of the website to scrape
url = 'https://www.public.fr/people/page/2'#/il-a-demonte-le-lavabo-les-toilettes-nagui-recoit-lenny-kravitz-tout-derape-dans-taratata'
url='https://www.public.fr/olivier-giroud-ses-freres-et-soeurs-persecutes-lex-joueur-de-lequipe-de-france-decide-dagir'
url='https://vsd.fr/actu-people/page/2/'

# Send an HTTP request to the URL
response = requests.get(url)

# Check if the request was successful
if response.status_code == 200:
    # Parse the content of the page
    soup = BeautifulSoup(response.content, 'html.parser')
    
    for link in soup.find_all('a'):
        # Check if the next element in the document is <h3>
        next_element = link.find_next()
        if next_element and next_element.name == 'h3':
            # Look for the preceding <img> element
            previous_img = link.find_previous('img')
            if previous_img and 'srcset' in previous_img.attrs:
                # Extract the 300w image from srcset
                srcset = previous_img['srcset']
                for img_entry in srcset.split(','):
                    if '300w' in img_entry:
                        image_url = img_entry.split(' ')[1]
                        break
                else:
                    image_url = "300w image not found"
                
                # Print the details
                title = next_element.get_text(strip=True)
                href = link.get('href')
                print(f"Title: {title}\nLink: {href}\n300w Image: {image_url}\n")
    resume=soup.find_all('p')
    print(resume[:2])
    print(soup)
else:
    print(f"Failed to retrieve the webpage: {response.status_code}")
```
